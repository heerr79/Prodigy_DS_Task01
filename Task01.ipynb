```python
#List of contries for the graph.
countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

#Filtering the data frame
filtered_data = df[df['Country Name'].isin(countries_to_plot)]

#Creating the bar chart
plt.bar (filtered_data['Country Name'], filtered_data['2022'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2022')
plt.title('Population of selected countries in 2022')
plt.xticks(rotation=45, ha='right') #Rotating the x-axis labels for readability
plt.show()
```

```python
#Extracting the required columns
countries = df['Country Name']
population_2022 = df['2022']

#sorting
sorted_data = df.sort_values(by='2022', ascending=False)

#top 15 countries of high population
top_n = 15
top_countries = sorted_data['Country Name'].head(top_n)

#creating a bar chart
plt.figure(figsize=(12, 7))
plt.bar(top_countries, sorted_data['2022'].head(top_n), color='purple')
plt.xlabel('Country')
plt.ylabel('Population in 2022')
plt.title(f'Distribution of Population in 2022 - Top {top_n} Countries')
plt.xticks(rotation=45)  
plt.show()
```

```python
# Extracting required columns 
countries = df['Country Name']
population_1996 = df['1997']

# List of countries 
selected_countries = ['China', 'India', 'Russia', 'Japan', 'Canada', 'Australia', 'Germany', 'France', 'Samoa', 'South Korea']

# Filtering the DataFrame for the selected countries
selected_data = df[df['Country Name'].isin(selected_countries)]

# Sorting
sorted_data = selected_data.sort_values(by='1997', ascending=False)

# Creating a bar chart
plt.figure(figsize=(11, 6))
plt.bar(sorted_data['Country Name'], sorted_data['1997'], color='pink')
plt.xlabel('Country')
plt.ylabel('Population in 1997')
plt.title(f'Distribution of Population of selected countries in 1997')
plt.xticks(rotation=45)
plt.show()
```
